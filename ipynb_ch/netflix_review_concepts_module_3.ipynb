## 第 1 節 — 變數、資料型別與基本運算子

### 變數與動態型別

在 Python 中,變數只是一個綁定到某個值的名稱——你不需要事先宣告它的型別,而且同一個名稱之後可以重新綁定到完全不同型別的值。這稱為**動態型別(dynamic typing)**。


````python
count = 10          # count is bound to an int
print(count, type(count))

count = "ten"       # now count is bound to a str -- perfectly legal
print(count, type(count))
````

### 核心資料型別

Python 內建的少數幾種型別經常出現:`int`、`float`、`str`、`bool`,以及特殊值 `None`。內建函式 `type()` 可以準確告訴你某個值是什麼型別。


````python
integer_value = 42
float_value = 3.14159
string_value = "hello"
boolean_value = True
nothing_value = None

print(type(integer_value))   # <class 'int'>
print(type(float_value))     # <class 'float'>
print(type(string_value))    # <class 'str'>
print(type(boolean_value))   # <class 'bool'>
print(type(nothing_value))   # <class 'NoneType'>
````

`None` 代表刻意表示「沒有值」。它和 `0`、`False` 或空字串是不同的,即使在布林情境中這些值全部都是「假值(falsy)」(第 2 節會深入介紹真假值判斷)。你會經常看到 `None` 被用作函式的預設參數(第 4 節),也會用來標示缺失的資料。

判斷型別時,`isinstance()` 通常是更好的工具,因為它也會考慮繼承關係(這在你讀到第 8 節的類別時會很重要):


````python
print(isinstance(integer_value, int))     # True
print(isinstance(True, int))              # True -- bool is a subclass of int in Python
````

最後一行是 Python 一個經典的怪癖:布林值在技術上是整數的子型別,這就是為什麼 `True + True` 會得到 `2` 而不會出錯。這在實務上很少造成影響,但它解釋了一些原本看起來令人困惑的行為。

### 基本運算子

算術運算的行為大致符合預期,其中有兩個運算子特別值得一提:`/` 一律執行**真除法(true division)**,回傳浮點數;而 `//` 執行**向下取整除法(floor division)**,會捨去餘數。


````python
a, b = 17, 5

print(a + b)    # 22   addition
print(a - b)    # 12   subtraction
print(a * b)    # 85   multiplication
print(a / b)    # 3.4  true division -- always returns a float
print(a // b)   # 3    floor division -- discards the remainder
print(a % b)    # 2    modulo -- the remainder itself
print(a ** b)   # 1419857  exponentiation
````

在該用 `//` 的地方誤用 `/`(或反過來),是後續在資料流程中計算陣列索引或批次大小時,常見卻不容易察覺的差一錯誤(off-by-one)來源——只要除法的結果會被拿去當索引使用,就值得多檢查一次。

字串串接使用相同的 `+` 運算子。f-string(第 6 節會更詳細介紹)是建立格式化輸出的標準、易讀寫法——只要在以 `f` 開頭的字串字面值中,直接把運算式放進 `{}` 即可。


````python
name = "Ada"
score = 97.5
print(f"{name} scored {score:.1f} points")   # Ada scored 97.5 points
````

### 銜接到 NumPy

Python 內建的數值型別很靈活,但用來處理大量數值資料時,速度和記憶體效率都不算好。`numpy`(以 `pip install numpy` 安裝,幾乎總是以社群標準慣例的別名 `np` 匯入)會把數字儲存在緊密排列、固定寬度的記憶體中,並且會對整個陣列一次套用相同的算術運算子(逐元素運算)。


````python
import numpy as np

values = np.array([1, 2, 3, 4, 5], dtype=np.float64)
print(values * 2)          # [ 2.  4.  6.  8. 10.]
print(values.dtype)        # float64
print(values.dtype.itemsize)  # 8 -- bytes per element
````

`dtype` 值得停下來多想一下:和一般的 Python 變數不同,NumPy 陣列中的每個元素都共用同一個固定、明確的型別。現在就熟悉 Python 原生型別和 NumPy 較嚴格、固定寬度型別之間的關係,能在之後遇到 `int` 與 `np.float64` 之間的型別不一致、並在訓練模型深處產生警告或非預期結果時,省下不少真正的困擾。

### 型別轉換

Python 使用型別本身作為函式來進行明確的型別轉換:`int(...)`、`float(...)`、`str(...)`。


````python
count_str = "42"
count_int = int(count_str)          # 42, as an actual integer
price_float = float("19.99")         # 19.99
back_to_str = str(count_int)          # "42"

print(count_int + 8)     # 50 -- now this works, because count_int is an int
````

這在從檔案或使用者輸入讀取資料時經常會遇到,因為從文字來源讀入的內容,預設一律是字串,即使它看起來像數字也一樣。忘記轉換(`"42" + 8` 會引發 `TypeError`)是最常見的早期錯誤之一——Python 不會自己默默猜測你到底是想做字串串接還是數值相加,修正方式幾乎都是像上面那樣明確地轉換型別。

### 增量賦值(Augmented assignment)

這是「根據變數目前的值來更新它」這種常見模式的簡寫寫法。


````python
total = 0
total += 5      # equivalent to: total = total + 5
total *= 2       # equivalent to: total = total * 2
print(total)     # 10
````

每個算術運算子都有對應的增量賦值寫法(`+=`、`-=`、`*=`、`/=`、`//=`、`**=`、`%=`),你會在迴圈中(第 3 節)經常看到它們,每次迭代用來更新累計總和、計數器或累加器。

還有一個值得在這裡先提一下的區別,因為它會在第 5 節再次出現:數字、字串和布林值在 Python 中都是**不可變(immutable)**的——`total += 1` 並不會就地修改那個整數,而是建立一個新的整數,再讓 `total` 重新綁定到它。對於簡單的值來說,這個差異是看不出來的,但對於串列(list)和字典(dictionary)這種可變(mutable)的物件來說,這就變成一個重要且可見的區別。

### 串聯比較(Chained comparisons)

Python 讓你可以用貼近日常數學記法、讀起來很自然的方式串聯多個比較。


````python
age = 25
print(0 <= age < 18)    # False -- equivalent to: 0 <= age and age < 18
print(18 <= age < 65)   # True
````

## 第 2 節 — 控制流程：條件判斷

### `if` 陳述式

Python 的條件判斷語法非常接近日常英語。只有第一個評估結果為 `True` 的分支會被執行。注意條件式外面沒有括號,程式區塊外面也沒有大括號——Python 改用冒號和一致的縮排來表示。這裡的縮排不只是風格問題:它決定了 Python 如何判斷哪些行屬於哪個區塊,所以縮排不一致會是貨真價實的語法錯誤。


````python
temperature = 72

if temperature > 80:
    print("It's hot")
elif temperature > 60:
    print("It's mild")
else:
    print("It's cold")
````

### 真假值判斷(Truthiness)

任何值都可以當作條件式使用,不是只有明確的 `True`/`False` 才行。`0`、`0.0`、`None`、空字串 `""`,以及空的容器(`[]`、`{}`、`()`)全部都是**假值(falsy)**;基本上其他所有東西都是**真值(truthy)**。


````python
data = []

if data:
    print("processing data")
else:
    print("no data to process")   # this branch runs -- an empty list is falsy
````

直接檢查一個容器(`if data:`),而不是比較它的長度是否為零(`if len(data) > 0:`),是符合 Python 慣例的寫法,這週你在閱讀程式碼時會經常看到這種寫法。

### 比較與邏輯運算子

比較運算子(`==`、`!=`、`<`、`>`、`<=`、`>=`)會產生布林值,你可以用 `and`、`or`、`not` 來組合它們。


````python
age = 25
has_license = True

if age >= 18 and has_license:
    print("eligible to drive")

if not has_license:
    print("cannot drive")   # doesn't print here, since has_license is True
````

有一個細節值得特別提出:`==` 檢查的是*值*是否相等,而 `is` 檢查的是*身分*——也就是兩個名稱是否指向記憶體中完全相同的物件。比較數字、字串以及大多數日常的值時,請使用 `==`。`is` 保留給更少數的情況使用,最常見的就是專門用來檢查是否為 `None`(`if result is None:`),這也是符合慣例、建議使用的寫法。

### 三元(條件)運算式

一種精簡的單行條件運算式,適合用在簡單的情況。


````python
x = 7
label = "even" if x % 2 == 0 else "odd"
print(label)   # odd
````

這等同於一個完整的 `if`/`else` 區塊,在各分支中對 `label` 進行賦值,只是濃縮成一行而已。一旦條件或分支變得複雜,它的可讀性明顯會變差——這時候,完整的 `if`/`else` 才是更好的選擇。

### 使用 `in` 進行成員檢查

一個非常常見的模式:檢查某個值是否出現在某個集合中。


````python
allowed_labels = ["cat", "dog", "bird"]
label = "fish"

if label in allowed_labels:
    print("recognized label")
else:
    print("unrecognized label")   # this branch runs
````

`in` 可以用在串列、tuple、字串(檢查是否包含子字串)上——正如第 5 節會提到的,在集合(set)上使用 `in` 特別快,因為集合正是為了這種成員檢查而設計的。

### 巢狀條件判斷


````python
income = 55000
has_dependents = True

if income > 50000:
    if has_dependents:
        tax_bracket = "B-with-credit"
    else:
        tax_bracket = "B"
else:
    tax_bracket = "A"

print(tax_bracket)   # B-with-credit
````

當巢狀結構超過兩、三層時,通常就是該重新整理邏輯的信號了——常見的做法是直接用 `and` 把條件組合起來,或是把內層邏輯抽出來獨立成一個函式(第 4 節)。過深的巢狀條件判斷是常見的可讀性問題,原因就在於每多一層,就更難一眼看出哪種條件組合會導向哪種結果。

### `match`/`case`

Python 3.10 以上版本新增了一種結構化模式比對(structural pattern matching)陳述式,作為在把某個值與多種可能性比較時,取代冗長 `if`/`elif` 鏈的另一種選擇。


````python
status_code = 404

match status_code:
    case 200:
        print("OK")
    case 404:
        print("Not Found")
    case 500:
        print("Server Error")
    case _:
        print("Unknown status")   # the underscore is a catch-all, like else
````

你會看到很多既有程式碼改用冗長的 `if`/`elif` 鏈來達成同樣的目的——兩者都是正確的寫法;`match`/`case` 只是針對這種特定情境,一種較新、有時可讀性更高的選項。

### 銜接到資料科學中的布林遮罩(boolean masks)

條件判斷在實務上最重要的延伸應用就是**布林索引(boolean indexing)**,你會在 NumPy 和 pandas 中經常用到它。與其為每個元素各寫一個明確的 `if`,你只需要寫一次條件式,它就會套用到整個陣列或表格上,產生一個布林值陣列,你可以用它來做篩選。


````python
import numpy as np
import pandas as pd

ages = np.array([15, 22, 34, 8, 61, 45])
is_adult = ages >= 18          # a boolean array, not a single True/False
print(is_adult)                # [False  True  True False  True  True]
print(ages[is_adult])          # [22 34 61 45] -- only the adult ages

df = pd.DataFrame({"age": ages, "name": ["A", "B", "C", "D", "E", "F"]})
adults_df = df[df["age"] >= 18]
print(adults_df)
````

當你想在陣列上組合一個以上的條件時,有一個重要的陷阱:Python 的 `and`/`or` 關鍵字只能用在單一布林值上,不能用在整個陣列上。要組合布林陣列,必須改用 `&` 和 `|` 運算子,並且把每個條件各自用括號括起來。


````python
young_adults = df[(df["age"] >= 18) & (df["age"] < 40)]
print(young_adults)
````

忘記加括號,或是習慣性地使用 `and`/`or`,是使用 pandas 篩選資料時最常見的早期錯誤之一——Python 會直接丟出錯誤,而不是默默做出錯誤的結果,但最好在遇到之前就先知道正確的寫法。

### 短路求值(Short-circuit evaluation)

`and` 和 `or` 不一定會對運算式的兩側都求值——只要整體結果已經確定,就會立刻停止。


````python
def is_valid(x):
    print(f"checking {x}")
    return x > 0

result = is_valid(-5) and is_valid(10)   # only "checking -5" prints
print(result)   # False
````

`is_valid(10)` 根本不會被執行,因為 `is_valid(-5)` 已經回傳 `False`,Python 知道整個 `and` 運算式無論如何都不可能是 True。這種短路特性經常被刻意當作防護機制使用:`if data and data[0] > 10:` 可以安全地先確認 `data` 不是空的,再嘗試存取 `data[0]`,而短路求值保證了當 `data` 為空時,那個有風險的存取動作永遠不會執行。

### `any()` 和 `all()`

用來在不寫明確迴圈的情況下,一次檢查整個集合是否符合某個條件。


````python
scores = [82, 91, 76, 88]

print(all(score >= 60 for score in scores))   # True -- everyone passed
print(any(score >= 90 for score in scores))   # True -- at least one A
````

`score >= 60 for score in scores` 是一個**產生器運算式(generator expression)**——就是一個沒有外層方括號的串列生成式(list comprehension),它是逐一產生值,而不是先在記憶體中建立完整的串列,`any()` 和 `all()` 可以在一旦知道答案時,就提早停止從中取值。


## 第 3 節 — 迴圈與迭代

### `for` 迴圈

Python 的 `for` 迴圈是直接對序列中的元素進行迭代,而不是自己手動維護一個索引計數器。


````python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(f"I like {fruit}")
````

當你需要一組特定的數字序列時,`range()` 可以產生這個序列,而不需要在記憶體中建立完整的串列。


````python
for i in range(5):        # 0, 1, 2, 3, 4
    print(i ** 2)
````

### `while` 迴圈

`for` 迴圈會迭代固定、已知的次數,而 `while` 迴圈則會在某個條件持續為真時繼續執行——當你事先不知道需要多少次迭代時很有用。


````python
value = 100
steps = 0
while value > 1:
    value = value / 2
    steps += 1
print(f"took {steps} steps to shrink below 1")
````

使用 `while` 迴圈時要特別小心:如果條件永遠不會變成假,迴圈就會無窮無盡地執行下去。相較於對固定大小的序列做 `for` 迴圈,這種錯誤更容易發生,所以每次寫 `while` 迴圈時都值得多加留意。

### `break` 與 `continue`

`break` 會立即跳出迴圈;`continue` 則會跳過目前這次迭代剩下的部分,直接進入下一次迭代。


````python
numbers = [4, 7, 2, 9, 3, 6]

for n in numbers:
    if n > 8:
        break            # stop looping entirely once we see a number > 8
    if n % 2 != 0:
        continue          # skip odd numbers, don't print them
    print(n)
````

### `enumerate` 與 `zip`

`enumerate()` 在迭代時同時給你索引和值,不需要自己手動維護一個計數器。


````python
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")
````

`zip()` 讓你可以同步迭代兩個(或多個)序列。


````python
names = ["Ada", "Grace", "Alan"]
scores = [97, 88, 91]

for name, score in zip(names, scores):
    print(f"{name}: {score}")
````

### 串列生成式(List comprehensions)

一種精簡的語法,可以透過對現有序列的每個元素套用某個運算式,來建立一個新的串列,途中也可以選擇性地進行篩選。


````python
squares = [n ** 2 for n in range(10)]
even_squares = [n ** 2 for n in range(10) if n % 2 == 0]
print(squares)
print(even_squares)
````

這和寫一個明確的 `for` 迴圈、不斷把元素附加到一個逐漸變長的串列中,做的事情完全一樣,只是寫法更精簡——這種寫法已經是慣用做法,即使你還沒有完全習慣自己動手寫,也應該要能夠讀懂它。

### 字典與集合生成式

同樣精簡的語法也可以延伸用來建立字典和集合。


````python
names = ["Ada", "Grace", "Alan"]
name_lengths = {name: len(name) for name in names}
print(name_lengths)          # {'Ada': 3, 'Grace': 5, 'Alan': 4}

unique_first_letters = {name[0] for name in names}
print(unique_first_letters)   # {'A', 'G'}
````

### 巢狀迴圈

迴圈可以彼此巢狀嵌套,這就是如何迭代兩個集合的每一種組合、或是網格中每一個格子的方法。


````python
rows = ["A", "B"]
columns = [1, 2, 3]

for row in rows:
    for column in columns:
        print(f"{row}{column}")   # A1, A2, A3, B1, B2, B3
````

巢狀迴圈是表達組合式迭代的自然方式,但成本會很快攀升——一個巢狀兩層、分別對大小為 $m$ 和 $n$ 的集合進行迭代的迴圈,其內層區塊會執行 $m \times n$ 次,而第三層巢狀又會讓這個次數再乘一次。每當你發現自己針對數值資料寫出雙層或三層巢狀迴圈時,都值得檢查看看是否可以改用 NumPy 運算來取代。

### 關於 `itertools` 的簡短說明

對於較不常見的迭代模式,標準函式庫的 `itertools` 模組(Python 內建,不需要 `pip install`)提供了高效、經過充分測試的工具,不需要你為了同樣的目的手動寫巢狀迴圈。


````python
from itertools import product

for row, column in product(["A", "B"], [1, 2, 3]):
    print(f"{row}{column}")   # same output as the nested loop above
````

### 為什麼明確的迴圈會被向量化運算取代

以下是這門課程實際上的重點所在。假設你想把一個很大的數值集合中的每個元素都平方,以下先用明確的 Python 迴圈來寫。


````python
import time

numbers_list = list(range(1_000_000))

start = time.time()
squared = [n ** 2 for n in numbers_list]
print(f"pure Python: {time.time() - start:.4f} seconds")
````

拿它和對應的 NumPy 運算比較一下。


````python
import numpy as np

numbers_array = np.arange(1_000_000)

start = time.time()
squared_np = numbers_array ** 2
print(f"NumPy vectorized: {time.time() - start:.4f} seconds")
````

執行這兩段程式碼,你通常會看到 NumPy 版本快上數十倍(實際數字依機器而異)。原因直接呼應了第 1 節的內容:Python 的 `for` 迴圈每次只處理一個 Python 物件,伴隨著這樣做所帶來的所有額外開銷,而 NumPy 的向量化運算則把整個計算下推到快速、經過編譯、固定寬度型別的程式碼中,一次對整個陣列進行運算。在整個實作課程中,每當你發現自己對 NumPy 陣列或 pandas 欄位寫出明確的 Python 迴圈時,都值得停下來想想是否有向量化運算可以做到同樣的事——通常是有的,而且寫起來通常更短,執行起來也快得多。

### 一個現在就該知道的 pandas 陷阱

pandas 的 DataFrame 支援 `.iterrows()`,讓你可以逐列迭代,就像迭代一個普通的串列一樣。


````python
import pandas as pd

df = pd.DataFrame({"x": [1, 2, 3, 4]})

for index, row in df.iterrows():
    print(index, row["x"])
````

這樣寫可以執行,但基於前面提到的完全相同的原因,這通常是處理 DataFrame 內容中最慢的方式。幾乎每一個看起來需要用 `.iterrows()` 的任務——根據現有欄位計算新欄位、篩選列、彙總數值——都有一個向量化的等價寫法(`df["x"] * 2`、第 2 節提到的布林遮罩、第 4 節的 `.apply()`,或是內建的彙總方法),執行起來會更快,寫起來通常也更短。把 `.iterrows()` 當成最後的手段,而不是預設習慣。


## 第 4 節 — 函式

函式讓你可以把一段程式碼包裝成一個可重複使用、有名稱的單元。你在這門課中已經用過的幾乎每一個工具——`LinearRegression().fit()`、`mean_squared_error()`、`np.cov()`——都是函式(或方法,也就是附加在物件上的函式,第 8 節會介紹)。

### 定義函式

`def` 用來引入一個函式,後面接著名稱、以括號括起來的參數列表,以及一個縮排區塊。`return` 會把一個值傳回給呼叫端,並立即結束函式。沒有明確 `return` 的函式,預設會回傳 `None`——這是忘記回傳原本想計算的值時,常見的困惑來源。


````python
def greet(name):
    return f"Hello, {name}!"

message = greet("Ada")
print(message)   # Hello, Ada!
````

### 參數與預設引數

參數可以有預設值,讓它們在呼叫時成為選填項目。


````python
def scale_value(x, factor=1.0, offset=0.0):
    return x * factor + offset

print(scale_value(10))                 # 10.0 -- uses both defaults
print(scale_value(10, factor=2.0))     # 20.0
print(scale_value(10, factor=2.0, offset=5.0))  # 25.0
````

這裡明確以 `factor=2.0`(**關鍵字引數,keyword argument**)呼叫,而不是單純傳入 `2.0`(**位置引數,positional argument**),雖然是選擇性的,但一旦函式有多個參數,這就會變成重要的可讀性慣例——你會在 scikit-learn 的 API 中到處看到這個慣例,幾乎每個第一個參數以外的引數,都預期以關鍵字方式傳入。

### `*args` 與 `**kwargs`

有時候你會希望一個函式可以接受不定數量的位置引數或關鍵字引數,而不需要事先為每一個命名。


````python
def summarize(*values, **labels):
    print(f"values: {values}")   # a tuple of all positional arguments
    print(f"labels: {labels}")   # a dict of all keyword arguments

summarize(1, 2, 3, unit="meters", precision=2)
````

你不會經常自己寫 `*args`/`**kwargs`,但你會在函式庫程式碼中經常看到它——這是許多 scikit-learn 和 pandas 函式用來把引數轉傳到更底層常式的方式——所以即使你很少自己寫,也要能夠認得這種語法。

### Docstring(文件字串)

放在函式內部最前面(在任何其他程式碼之前)的字串字面值,會成為它的**docstring**——也就是工具、IDE 和內建的 `help()` 函式都能自動取得的說明文件。


````python
def z_score(x, mean, std):
    '''Standardize a value using the given mean and standard deviation.

    Returns (x - mean) / std.
    '''
    return (x - mean) / std

help(z_score)
````

養成習慣,為任何用途無法從名稱一眼看出的函式,至少寫一行 docstring。

### 作用域(Scope)

在函式內建立的變數是該函式的**區域變數(local)**——它在函式外不存在,也不會覆寫其他地方定義的同名變數。


````python
total = 100

def reset_total():
    total = 0        # this creates a new, local "total" -- it does NOT change the outer one
    return total

print(reset_total())   # 0
print(total)            # 100 -- unchanged
````

這是一項刻意設計的安全機制:函式內部的變數不會不小心破壞掉呼叫它的程式碼中的變數。

### `global` 關鍵字

有時候你確實會想讓一個函式修改在它外部定義的變數,而不是建立一個新的區域變數。Python 要求你必須明確這樣宣告。


````python
total = 100

def add_to_total(amount):
    global total
    total += amount

add_to_total(50)
print(total)   # 150 -- the outer variable really was changed this time
````

實務上,依賴 `global` 通常是一個訊號,代表這個函式應該改成把它需要的值作為參數傳入,並回傳更新後的結果,而不是伸手到外部去修改共用的狀態——只讀取和寫入自己參數與區域變數的程式碼,更容易獨立理解和測試。

### 型別提示(Type hints)

Python 仍然是動態型別語言,但你可以選擇性地為函式預期的參數與回傳型別加上註記,作為文件說明,工具和 IDE 可以幫你檢查,而不會改變程式碼實際的執行方式(第 9 節會介紹一個真正會檢查這些註記的工具 `mypy`)。


````python
def z_score(x: float, mean: float, std: float) -> float:
    return (x - mean) / std

print(z_score(10.0, 5.0, 2.0))
````

Python 預設在執行期並不會強制檢查這些型別註記——傳入字串給標註為 `float` 的參數,本身不會引發錯誤——但它們讓函式預期的輸入與輸出具備自我說明的效果。

### 遞迴函式

函式可以呼叫自己,這種技巧稱為**遞迴(recursion)**,最適合用在可以拆解成自身較小版本的問題上。


````python
def factorial(n):
    if n <= 1:
        return 1                     # base case -- stops the recursion
    return n * factorial(n - 1)      # recursive case -- calls itself

print(factorial(5))   # 120
````

每一個遞迴函式都需要一個**基準情況(base case)**——一個直接回傳值、不再呼叫自己的條件——否則它會無窮遞迴下去(最終會因為 `RecursionError` 而崩潰)。

### 一個常見陷阱:可變的預設引數

這一點值得特別提出來,因為第一次遇到時,行為會讓人相當意外。預設引數的值只會在函式*被定義*的當下求值一次——而不是每次呼叫時都求值——如果這個預設值剛好是像串列這樣的可變物件,就會造成問題。


````python
def add_item(item, collection=[]):     # BUG: the same list is reused every call
    collection.append(item)
    return collection

print(add_item("a"))   # ['a']
print(add_item("b"))   # ['a', 'b'] -- surprising! it's the SAME list as before
````

標準的修正方式是把預設值設為 `None`,並在需要時,於函式內部建立一個新的空集合。


````python
def add_item_fixed(item, collection=None):
    if collection is None:
        collection = []
    collection.append(item)
    return collection

print(add_item_fixed("a"))   # ['a']
print(add_item_fixed("b"))   # ['b'] -- a fresh list each time, as expected
````

這正是第 2 節提到的慣用 `is None` 檢查方式,也是 Python 中最常被提到的「陷阱」之一,而且原因十足充分——這個錯誤很容易被引入,而且因為只有在多次呼叫之後才會顯現,快速測試時很容易被忽略。

### Lambda 函式

對於簡短、用完即丟的函式,Python 提供了一種內嵌語法,完全省略 `def` 關鍵字和函式名稱。


````python
square = lambda x: x ** 2
print(square(5))   # 25
````

### 函式作為第一級值(first-class values)

在 Python 中,函式和其他值一樣:你可以把它當作引數傳給另一個函式、儲存在變數中,或放進串列裡。這正是這週你會用到的最常見資料科學模式之一——對 pandas DataFrame 的每一列或每一欄套用一個函式——背後的機制。


````python
import pandas as pd

df = pd.DataFrame({"celsius": [0, 20, 37, 100]})

def to_fahrenheit(c):
    return c * 9 / 5 + 32

df["fahrenheit"] = df["celsius"].apply(to_fahrenheit)
print(df)
````

`.apply()` 接收的是你的函式本身——而不是呼叫它之後的結果——並且會對每個元素各呼叫它一次,做的事情就跟第 3 節的 `for` 迴圈一樣,只是寫法精簡許多。用 lambda 取代單獨定義的函式,套用同樣的模式,在非常簡短的轉換中很常見。


````python
df["fahrenheit_lambda"] = df["celsius"].apply(lambda c: c * 9 / 5 + 32)
print(df)
````

認識到 `.apply()`、`.map()` 這類 pandas/NumPy 方法,本質上就是「對每個元素呼叫一次這個函式」——和迴圈是同樣的概念,只是交給一個能更快執行的函式庫來處理——會讓你這週遇到的大部分資料處理程式碼,感覺不再那麼神祕。


## 第 5 節 — 核心資料結構:串列、Tuple、字典與集合

### 串列(Lists)

串列是一種有序、可變(可修改)的集合,以方括號表示。


````python
scores = [85, 92, 78, 90]

scores.append(88)          # add an element to the end
scores[0] = 87              # change an existing element by index
print(scores[-1])           # 88 -- negative indices count from the end
print(scores[1:3])          # [92, 78] -- slicing: elements at index 1 and 2
print(scores)
````

串列是 Python 中預設、通用型的集合,經常用來累積結果、保存一系列混合或單一型別的資料,以及在你的程式之間傳遞集合資料。

### Tuple

Tuple 看起來和串列類似,但它是**不可變(immutable)**的——一旦建立,其內容就不能再更改——並以括號表示(通常也常常完全不加任何括號)。


````python
point = (3, 4)
x, y = point                # unpacking: assigns 3 to x and 4 to y in one line
print(x, y)                 # 3 4
````

不可變性是一種特性,而不是限制:它表示這個值本來就是要保持固定不變的,而且這讓 tuple 可以作為字典的鍵(串列由於是可變的,不能作為字典的鍵)。Tuple 也常被用作函式的回傳值,當函式需要一次回傳一個以上的結果時。


````python
def min_and_max(values):
    return min(values), max(values)   # returns a tuple

lowest, highest = min_and_max([4, 8, 1, 9, 3])
print(lowest, highest)    # 1 9
````

### 字典(Dictionaries)

字典儲存鍵值對(key-value pairs),讓你可以透過鍵來快速查找,而不是透過數字位置。


````python
student = {"name": "Ada", "age": 28, "major": "Computer Science"}

print(student["name"])        # Ada
student["age"] = 29            # update an existing key
student["gpa"] = 3.9            # add a new key

for key, value in student.items():
    print(f"{key}: {value}")
````

字典是表示單一結構化記錄——也就是一筆帶有具名欄位的資料——的自然方式,而字典所組成的串列,是真實世界資料在載入到像 pandas DataFrame 這樣更專門的結構之前,最常見的呈現方式之一,你在下面就會看到。

### 集合(Sets)

集合是一組無序且元素唯一的集合——當你關心的是成員資格或唯一性,而不是順序或位置時,就很有用。


````python
tags_a = {"python", "data", "ml"}
tags_b = {"ml", "stats", "python"}

print(tags_a | tags_b)     # union: {'python', 'data', 'ml', 'stats'}
print(tags_a & tags_b)     # intersection: {'python', 'ml'}
print(tags_a - tags_b)     # difference: {'data'}

print("python" in tags_a)  # True -- membership testing is very fast on a set
````

### 選擇合適的資料結構

一個快速的判斷方式:當順序重要、且內容可能會改變時使用**串列(list)**;當值是固定不變的時候使用**tuple**;當你需要用有意義的名稱、而不是位置來查找值時使用**字典(dictionary)**;當你關心唯一性或快速成員檢查、而不在意順序時使用**集合(set)**。

### 排序

`sorted()` 可以用在以上任何一種集合上,並且接受一個 `key` 引數——一個函式,這正好呼應了第 4 節提到函式是普通值的觀點——告訴它要依據什麼來排序。


````python
students = [
    {"name": "Ada", "score": 91},
    {"name": "Grace", "score": 88},
    {"name": "Alan", "score": 95},
]

by_score = sorted(students, key=lambda s: s["score"], reverse=True)
for s in by_score:
    print(s["name"], s["score"])
# Alan 95
# Ada 91
# Grace 88
````

`sorted()` 會回傳一個新的、已排序的串列,而不會改變原本的串列;串列也有一個就地排序的 `.sort()` 方法,會直接修改該串列並回傳 `None`——一個常見的小錯誤是寫成 `students = students.sort()`,這會讓 `students` 變成 `None`,因為 `.sort()` 不像 `sorted()` 那樣會回傳排序後的串列。

### 巢狀結構

這四種集合型別可以自由組合:字典組成的串列、串列組成的字典、值本身又是字典的字典,依此類推。真實世界的結構化資料,尤其是從 JSON 載入的資料,經常會巢狀嵌套兩到三層深。


````python
course = {
    "name": "Data Science Intensive",
    "students": [
        {"name": "Ada", "scores": [91, 88, 95]},
        {"name": "Grace", "scores": [88, 90, 85]},
    ],
}

for student in course["students"]:
    average = sum(student["scores"]) / len(student["scores"])
    print(f"{student['name']}: {average:.1f}")
````

閱讀像這樣的巢狀結構,主要就是由外而內、一次處理一層 `[]` 或 `.key` 存取——正如上面迴圈中的模式那樣,先進入 `course["students"]`,再進入每一個 `student["scores"]`。

### 關於複製可變結構的說明

因為串列和字典都是可變的,把其中一個賦值給一個新名稱,並不會建立一份獨立的複本——兩個名稱指向的是完全相同的底層物件。


````python
original = [1, 2, 3]
alias = original          # alias is NOT a copy -- it's the same list
alias.append(4)
print(original)            # [1, 2, 3, 4] -- the "original" changed too!
````

要取得真正獨立的複本,請使用 `.copy()`(或是 `list()`/`dict()` 建構函式),但要記住,這只複製外層結構——也就是**淺複製(shallow copy)**。如果結構中包含巢狀的可變物件(例如串列組成的串列),除非改用 `copy.deepcopy()`,否則那些內層物件仍會在原本的物件與複本之間共用。這個細微之處讓經驗豐富的程式設計師和初學者一樣容易踩雷——每當你不確定某個操作是否會修改到你不想更動的資料時,就值得像上面這樣直接測試看看。

### 銜接到資料科學結構

字典組成的串列——每個字典代表一筆記錄,擁有同一組鍵——可以直接轉換成 pandas DataFrame。


````python
import pandas as pd

records = [
    {"name": "Ada", "age": 28, "score": 91},
    {"name": "Grace", "age": 34, "score": 88},
    {"name": "Alan", "age": 41, "score": 95},
]

df = pd.DataFrame(records)
print(df)
````

同樣地,一個由數字組成的普通 Python 串列,可以直接轉換成 NumPy 陣列——正如第 1 節提到的,這會把相同的數字打包進更緊湊、固定型別、速度快得多的結構中。


````python
import numpy as np

plain_list = [4, 8, 15, 16, 23, 42]
array_version = np.array(plain_list)
print(array_version.mean())   # 18.0
````

字典也經常被用來在傳遞給模型之前,保存設定值或超參數(hyperparameters)。


````python
from sklearn.linear_model import Ridge

hyperparameters = {"alpha": 1.0, "fit_intercept": True, "random_state": 0}
model = Ridge(**hyperparameters)   # unpack the dict directly into keyword arguments
print(model)
````

`**hyperparameters` 這個語法,是第 4 節看到的 `**kwargs` 在字典上的對應版本——它會把字典的鍵值對直接解包成函式呼叫的關鍵字引數,這正是上面那一行如何用單獨一行、容易修改的程式碼,把 `alpha=1.0, fit_intercept=True, random_state=0` 傳給 `Ridge` 的方式。


## 第 6 節 — 字串、檔案 I/O 與錯誤處理

### 字串基礎

字串支援索引和切片,就跟串列一樣,還附帶大量內建方法,用於常見的轉換操作。


````python
text = "Data Science"

print(text.lower())          # data science
print(text.upper())          # DATA SCIENCE
print(text.split())          # ['Data', 'Science']
print(text.replace("Data", "Computer"))   # Computer Science
print(text[0:4])             # Data -- slicing works the same as with lists
print(len(text))             # 12
````

還有幾個常用到、值得現在就認識的字串方法:`.strip()` 會移除開頭和結尾的空白字元(在清理從檔案讀入的文字時極為常見,因為這類文字經常帶有多餘的換行字元),而 `.join()` 則是 `.split()` 的反向操作,把一個字串串列合併成一個字串,並在中間插入指定的分隔符號。


````python
messy = "   raw input line \n"
clean = messy.strip()
print(repr(clean))     # 'raw input line'  -- note: repr() shows quotes and escapes

words = ["Data", "Science", "Intensive"]
joined = "-".join(words)
print(joined)           # Data-Science-Intensive
````

### f-string

### 錯誤處理

事情總會出錯——檔案可能不存在,某個值可能無法轉換成你預期的數字,串列索引可能超出範圍。Python 透過丟出**例外(exceptions)**來標示這些問題,你用 `try`/`except` 來處理它們。


````python
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        print("cannot divide by zero, returning None instead")
        return None

print(safe_divide(10, 2))   # 5.0
print(safe_divide(10, 0))   # prints a message, returns None
````

捕捉特定的例外型別(上面的 `ZeroDivisionError`),而不是捕捉所有可能的錯誤,是建議的做法——一個空的 `except:` 會默默吞掉你沒有預料到的錯誤,包括真正的程式錯誤(bug),這會讓之後追查問題變得困難許多。

### 多個 `except` 子句和 `finally`

一個 `try` 區塊可以處理好幾種不同類型的錯誤,各自使用自己的 `except` 子句,還可以加上一個選擇性的 `finally` 區塊,無論如何都會執行。


````python
def load_and_parse(path):
    try:
        with open(path, "r") as f:
            contents = f.read()
        return int(contents.strip())
    except FileNotFoundError:
        print(f"file not found: {path}")
        return None
    except ValueError:
        print(f"file did not contain a valid integer")
        return None
    finally:
        print("finished attempting to load the file")   # always runs

# demonstrate all three paths: missing file, bad content, and success
load_and_parse("does_not_exist.txt")

with open("bad_number.txt", "w") as f:
    f.write("not a number")
load_and_parse("bad_number.txt")

with open("good_number.txt", "w") as f:
    f.write("42")
print(load_and_parse("good_number.txt"))
````

這裡的順序很重要:Python 會依序檢查 `except` 子句,並使用第一個符合實際丟出之例外的子句,所以如果同時有較特定和較一般的例外型別,請把較特定的放在前面。

### 拋出自己的例外

你也可以拋出自己的例外,用來表示某個函式收到了它無法妥善處理的輸入。


````python
def load_scores(values):
    if not values:
        raise ValueError("values list cannot be empty")
    return sum(values) / len(values)

try:
    load_scores([])
except ValueError as e:
    print(f"caught an error: {e}")
````

### 自訂例外類別

對於較大型的程式,常見的做法是定義自己的例外型別——這也預告了第 8 節會深入介紹的類別語法——這樣呼叫端就能區分你特定的錯誤情況和 Python 通用的內建錯誤。


````python
class InsufficientDataError(Exception):
    '''Raised when there isn't enough data to compute a reliable result.'''
    pass

def compute_average(values, minimum_count=5):
    if len(values) < minimum_count:
        raise InsufficientDataError(
            f"need at least {minimum_count} values, got {len(values)}"
        )
    return sum(values) / len(values)

try:
    compute_average([1, 2])
except InsufficientDataError as e:
    print(f"caught a custom error: {e}")
````

你會在之後的實作課程中載入外部資料時,經常用到完全一樣的模式:把載入步驟包在 `try`/`except` 中,並在檔案缺失、格式錯誤,或不符合你程式碼預期時,明確且大聲地失敗(而不是默默產生錯誤的結果)。

## 第 8 節 — 物件導向程式設計

物件導向程式設計(Object-oriented programming,OOP)把資料和操作這些資料的函式,一起打包成一個單元,稱為**物件(object)**。這門課中你早就一直在使用物件了,每次你對一個 scikit-learn 模型呼叫 `.fit()`,或是對一個 NumPy 陣列呼叫 `.mean()`,都是在使用物件。這一節會告訴你如何建立自己的物件。

### 類別與實例

**類別(class)**是建立物件的藍圖;**實例(instance)**則是根據這個藍圖建立出來的一個具體物件。


````python
class Dog:
    def __init__(self, name, breed):
        self.name = name      # an attribute -- data belonging to this instance
        self.breed = breed

    def bark(self):
        return f"{self.name} says Woof!"

my_dog = Dog("Rex", "Labrador")     # creates an instance
print(my_dog.name)                   # Rex
print(my_dog.bark())                 # Rex says Woof!
````

`__init__` 是一個特殊方法——稱為**建構子(constructor)**——會在建立新實例時自動執行,負責設定該實例的初始屬性。`self` 指的是這個方法被呼叫時所對應的那個特定實例;它永遠是方法的第一個參數,而且 Python 會自動傳入它——你在呼叫 `my_dog.bark()` 時不需要自己提供它。

### 實例屬性與類別屬性

在 `__init__` 內部賦值給 `self` 的每一樣東西,都是**實例屬性(instance attribute)**——每一個實例都有自己獨立的一份。而**類別屬性(class attribute)**則是直接定義在類別內部(而不是在某個方法內),由該類別的所有實例共用。


````python
class Dog:
    species = "Canis familiaris"     # class attribute -- shared by all dogs

    def __init__(self, name, breed):
        self.name = name              # instance attribute -- unique per dog
        self.breed = breed

rex = Dog("Rex", "Labrador")
fido = Dog("Fido", "Poodle")

print(rex.species, fido.species)   # Canis familiaris Canis familiaris -- shared
print(rex.name, fido.name)          # Rex Fido -- distinct per instance
````

對於真正在所有實例之間都相同的值(常數、預設設定、共用計數器),類別屬性是正確的選擇;任何特定於某一個物件的內容,則應該放在 `__init__` 中,作為實例屬性。

關於封裝的說明:Python 不像某些語言那樣真正強制私有屬性,但它有一個強烈的命名慣例——單一開頭底線(`self._internal_value`)表示「這是實作細節,請不要從類別外部直接存取它」,即使技術上沒有任何東西真的阻止你這麼做。你會在 scikit-learn 的原始碼中到處看到這個相同的慣例。

### 繼承(Inheritance)

一個類別可以被建立成另一個類別的特化版本,繼承它的屬性和方法,並增加或覆寫特定的部分。


````python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} makes a sound"

class Cat(Animal):                   # Cat inherits from Animal
    def speak(self):                  # override: cats speak differently
        return f"{self.name} says Meow!"

generic_animal = Animal("Creature")
my_cat = Cat("Whiskers")

print(generic_animal.speak())   # Creature makes a sound
print(my_cat.speak())            # Whiskers says Meow!
````

這種模式——一個共用的基底類別定義共同介面,由特定的子類別提供各自的方法實作——稱為**多型(polymorphism)**,這正是 scikit-learn 一致性設計背後的原理:`LinearRegression`、`Ridge`,以及其他每一個 scikit-learn 模型都共用相同的基底介面(`fit()`、`predict()`),但每一個在內部實際實作這些方法的方式都非常不同。

### Dunder 方法

以雙底線包住的方法,像 `__init__`,稱為**dunder 方法(dunder methods,double underscore 的縮寫)**,Python 會在特定情況下自動呼叫它們,而不是由你直接呼叫。


````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        return f"Point({self.x}, {self.y})"    # controls how print() displays it

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y   # controls ==

p1 = Point(1, 2)
p2 = Point(1, 2)

print(p1)             # Point(1, 2) -- uses __repr__, not a generic object address
print(p1 == p2)        # True -- uses __eq__, comparing values, not object identity
````

如果沒有 `__repr__`,印出一個實例時會顯示像 `<__main__.Point object at 0x7f...>` 這樣沒什麼幫助的內容;如果沒有 `__eq__`,`==` 就會退回身分比較(這兩個是否真的是記憶體中同一個物件?),所以兩個座標相同、但分別建立的點物件比較起來會是不相等的。對於任何你寫來代表簡單數值的類別來說,定義這兩個 dunder 方法都是常見的做法。

### 建立你自己的 scikit-learn 風格估計器

為了讓這個連結完全具體化,以下是一個小型類別,模仿你在這門課中一直使用的 `fit()`/`predict()` 介面,用統計模組中的常態方程式(normal-equations)公式,親手實作普通最小平方法(OLS)。


````python
import numpy as np

class SimpleLinearModel:
    def __init__(self):
        self.beta_ = None      # trailing underscore: a scikit-learn convention
                                 # for attributes only set after fitting

    def fit(self, X, y):
        X_design = np.column_stack([np.ones(len(X)), X])
        self.beta_ = np.linalg.inv(X_design.T @ X_design) @ X_design.T @ y
        return self             # returning self allows method chaining

    def predict(self, X):
        X_design = np.column_stack([np.ones(len(X)), X])
        return X_design @ self.beta_

rng = np.random.default_rng(0)
x = rng.uniform(0, 10, 50)
y = 2.5 * x + 3.0 + rng.normal(0, 2.0, 50)

model = SimpleLinearModel()
model.fit(x, y)
predictions = model.predict(x)
print(f"fitted parameters: {model.beta_}")
````

這裡的每一個部分看起來都應該很眼熟:`__init__` 設定物件的初始狀態,`fit()` 把它學到的東西以屬性的形式儲存在 `self` 上(使用 scikit-learn 自己也是為了同樣目的而遵循的、結尾加底線的命名慣例),而 `predict()` 則使用那些儲存下來的屬性來產生新的輸出。親手寫這個類別,應該能讓 `LinearRegression().fit(X, y)` 內部發生的事情,感覺不再那麼像個黑盒子——它其實就是同樣的常態方程式運算,只是被包裝成一個具有標準化介面、可重複使用的物件而已。

### Linter:在你執行程式碼之前就先抓出錯誤

**Linter** 執行的是靜態分析——在不執行程式碼的情況下閱讀它——藉此標示出可能的錯誤、未使用的程式碼,以及違反風格規範之處。`ruff` 是一個快速、廣泛使用的 linter。


````python
%%writefile bad_example.py
import numpy as np
import pandas as pd

def compute_average(values):
    total = 0
    for i in range(len(values)):
        total = total + values[i]
    average = total / len(values)
    return average

result = compute_average([1, 2, 3, 4, 5])
print(result)
````

````python
!python3 -m ruff check bad_example.py
````

這兩個未使用的匯入(`numpy`、`pandas`)在實際執行這個腳本時都不會造成錯誤——Python 不在意你是否匯入了從未使用過的東西。而這正是 linter 存在的意義:它能抓出那些技術上合法、但幾乎可以肯定是錯誤的地方(忘了刪除的匯入、賦值後從未讀取的變數、恆為真的比較),讓這些問題在累積成真正令人困惑、搞不清楚一個檔案到底依賴什麼之前,就先被抓出來。

### 型別檢查工具

### 自動化(單元)測試

Linter 和型別檢查工具抓的是形式上的錯誤;它們無法告訴你你的函式算出來的答案是不是對的。這正是**測試(test)**的工作:一小段程式碼,用已知的輸入呼叫你的函式,並把結果和已知的正確輸出做比對。`pytest` 是 Python 中這方面的標準工具。


````python
%%writefile stats_utils.py
def z_score(x, mean, std):
    if std == 0:
        raise ValueError("standard deviation cannot be zero")
    return (x - mean) / std
````

````python
%%writefile test_stats_utils.py
from stats_utils import z_score
import pytest

def test_z_score_basic():
    assert z_score(10, 5, 2) == 2.5

def test_z_score_zero_mean():
    assert z_score(0, 0, 1) == 0.0

def test_z_score_raises_on_zero_std():
    with pytest.raises(ValueError):
        z_score(10, 5, 0)
````

````python
!python3 -m pytest test_stats_utils.py -v
````

注意第三個測試:它刻意檢查用不良輸入(`std=0`)呼叫函式時,是否會如預期般丟出例外,使用的是 `pytest.raises`——一個好的測試套件不只驗證成功的情況,也會驗證失敗的行為。像這樣的測試套件真正的回報,會在你第二次動到這個函式時出現:如果你下個月重構 `z_score`,不小心破壞了對標準差為零的檢查,這三個測試會立刻抓到問題,而不是讓這個 bug 默默地存在,幾週後才以下游某處出現錯誤數字的形式浮現。測試套件實際上就是一張安全網,讓你可以放心地修改程式碼,而不用害怕動到它。

一個相關的工具 `coverage.py`(`pip install coverage`)會回報你的測試套件實際執行到你程式碼的比例——這是一個很有用的方式,可以找出完全沒有測試覆蓋到的函式或分支(一個你從未觸發過的 `except` 子句、一個你從未讓它為假的 `if` 條件)。高覆蓋率不保證測試品質良好,但低覆蓋率可以可靠地標示出沒有人驗證過的程式碼。

### 向 AI 程式設計工具尋求評論

像 Claude、GitHub Copilot 這類的 AI 助理工具,越來越常被納入正常的程式碼審查流程中——不是要取代上述工具,而是作為額外的一名審查者,能夠閱讀整個函式或檔案,並用自然語言進行推理。它們特別擅長幾個特定的問題:「這個函式沒有處理哪些邊界情況?」、「有沒有更簡單的寫法?」,以及「這個 docstring 真的符合程式碼實際的行為嗎?」。以下幾個習慣可以讓這件事真正變得有用,而不只是產生一堆聽起來煞有其事的雜訊:

- 提出具體的問題,而不是「這樣寫好嗎?」——「如果 `values` 是空的會發生什麼事?」比起籠統地要求一般性回饋,能得到更有行動力的答案。
- 把每一個建議都當成一個需要驗證的說法,而不是一個必須照做的指令。對結果執行 linter 和測試套件;AI 工具可能會建議一個讀起來更符合慣例、但悄悄改變了行為的變更,而本節前面提到的工具正是用來抓出這種問題的。
- 把它用在靜態工具在結構上無法檢查的事情上——變數名稱是否真的傳達了它的用途、一個函式是否同時做了太多不相關的事情、docstring 是否已經和它描述的程式碼脫節。Linter 和型別檢查工具精確但狹隘;AI 審查者的價值在於廣度和判斷力,代價是偶爾會自信滿滿地說錯——這正是為什麼上面的驗證步驟很重要。

### 程式碼審查與結對程式設計

同樣的原則也適用於人類協作者:在程式碼合併之前讓別人先讀過,能抓到任何自動化工具都抓不到的另一類問題——不清楚的命名、之後很難擴充的設計、一個現在成立但下一季就不成立的假設。一個有用的準則,不論審查者是人還是 AI 工具,就是優先審查行為和清晰度(這段程式碼真的做到它宣稱的事嗎?下一個人能看懂嗎?),而不是風格,因為風格早已由格式化工具自動處理掉了。

### 防禦性程式設計與斷言(assertions)

第 6 節介紹過針對真正無效的輸入拋出例外。一個較輕量、相關的工具是 `assert` 陳述式,通常用來抓程式設計者自己的錯誤——也就是如果其餘程式碼都正確,邏輯上應該不可能發生的狀態——而不是用來驗證外部輸入。


````python
def train_test_split_ratio(n_total, n_train):
    assert 0 < n_train < n_total, "n_train must be between 0 and n_total"
    return n_train / n_total

print(train_test_split_ratio(100, 70))

try:
    train_test_split_ratio(100, 150)
except AssertionError as e:
    print(f"assertion failed: {e}")
````

和針對不良外部資料而寫的 `raise ValueError(...)` 不同,`assert` 的用意是作為你自己程式碼內部不變條件(invariant)的健全性檢查,而且在最佳化過的正式環境執行中,常常會被整個移除——這正是為什麼它絕對不應該被用來驗證來自檔案、API 或使用者的、真正不可信的輸入,在這種情況下,第 6 節的 `try`/`except` 和明確的 `raise` 才是正確的工具。

### 讓函式保持精簡並避免重複

有兩個密切相關的習慣,其回報幾乎超過本節中任何單一工具:一個只做一件可以清楚命名之事情的函式,遠比一個同時做五件事的函式更容易測試、審查和重複使用;而同樣幾行程式碼在三個地方各複製貼上一次(而不是抽出來變成一個共用函式),意味著之後要找出並修正同一個 bug 時,得跑三個不同的地方——這個原則通常被概括為 **DRY:不要重複自己(don't repeat yourself)**。這一節裡沒有任何工具會強制執行這兩個習慣;兩者都是需要靠經驗慢慢掌握的判斷,而它們通常正是一次好的程式碼審查或 AI 評論最先會指出的地方。

### 版本控制的紀律

最後,還有一個超越單一檔案範圍的習慣:以小而聚焦的變更、搭配清楚的訊息來提交你的工作,能讓這一節裡的每一項工具搭配起來都更有用。比較一下:

```bash
git commit -m "fix off-by-one error in train/test split"   # 好:具體、意圖清楚
git commit -m "updates"                                     # 差:之後什麼都看不出來
```

當一次失敗的測試或 linter 警告,是由一個小而描述清楚的變更所引入,而不是埋藏在一次觸及數十個檔案的巨大提交中時,追查起原因來會容易得多。
