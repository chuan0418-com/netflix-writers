## 練習 10 — 打造爆紅分數

```python
df = netflix.copy()

df["hit_score"] = (
    df["netflix_viewing_hours"] / df["netflix_viewing_hours"].max()
)

df.sort_values("hit_score", ascending=False).head(10).rename(columns=COLUMN_NAMES_ZH)
```
