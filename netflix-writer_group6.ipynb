## Check match coverage


```python
import numpy as np
import pandas as pd

def match_coverage(frame: pd.DataFrame, source_columns: dict[str, str]) -> pd.DataFrame:
    """Report what fraction of rows have a non-null match for each source.

    Args:
        frame: the composite dataset.
        source_columns: maps a human-readable source name to one column
            from that source to check for nulls, e.g. {"tmdb": "tmdb_title"}.

    Returns:
        A small DataFrame with one row per source: matched count, total
        count, and coverage as a percentage.
    """
    rows = []
    total = len(frame)
    for source_name, column in source_columns.items():
        matched = frame[column].notna().sum()
        rows.append({"source": source_name, "matched": matched, "total": total,
                      "coverage_pct": round(100 * matched / total, 1)})
    return pd.DataFrame(rows)


match_coverage(df, {"tmdb": "tmdb_title", "imdb": "imdb_title"})
```

## Define the target: what counts as a "hit"?


```python
def label_hits(frame, source_cols=("netflix_viewing_hours", "imdb_numVotes", "tmdb_popularity"),
               quantile=0.75, min_sources=2):
    votes = pd.DataFrame({
        col: frame[col] >= frame[col].quantile(quantile) for col in source_cols
    })
    return votes.sum(axis=1) >= min_sources

df["is_hit"] = label_hits(df)
df["is_hit"].value_counts()


```

```python

# Validate the design of label_hits(): why 3 sources, why min_sources=2,
# and why quantile=0.75?
hit_source_cols = ["netflix_viewing_hours", "imdb_numVotes", "tmdb_popularity"]

# 1) Missingness: if we required all 3 sources to agree, how many rows would
#    even be eligible?
print("=== Missing rate per source ===")
print(df[hit_source_cols].isna().mean().round(3))
print()
print("all 3 sources present:", df[hit_source_cols].notna().all(axis=1).sum(), "/", len(df))
print("at least 2 sources present:", (df[hit_source_cols].notna().sum(axis=1) >= 2).sum(), "/", len(df))
print()

# 2) Correlation between sources: are they redundant, or do they capture
#    different things (which justifies cross-checking across sources)?
print("=== Spearman correlation between the 3 hit-signal sources ===")
print(df[hit_source_cols].corr(method="spearman").round(2))
print()

# 3) Sensitivity: how does the hit rate change across quantile / min_sources
#    combinations? Confirms quantile=0.75, min_sources=2 sits in a reasonable
#    middle ground (not too lenient, not too strict / small-sample).
print("=== Hit rate by quantile x min_sources ===")
for q in [0.5, 0.75, 0.9]:
    for m in [1, 2, 3]:
        n = label_hits(df, quantile=q, min_sources=m).sum()
        print(f"quantile={q:.2f}  min_sources={m}  ->  hits={n:3d}  ({100 * n / len(df):.1f}%)")

```

## Binge velocity: hours viewed per week on the list


```python
def binge_velocity(frame: pd.DataFrame,
                    hours_col: str = "netflix_viewing_hours",
                    weeks_col: str = "netflix_weeks") -> pd.Series:
    """Average viewing hours per week on the Top 10 list.

    Guards against division by zero for any title with weeks_col == 0.
    """
    weeks = frame[weeks_col].replace(0, np.nan)
    return (frame[hours_col] / weeks).rename("binge_velocity")


df["binge_velocity"] = binge_velocity(df)
df[["netflix_title", "netflix_viewing_hours", "netflix_weeks", "binge_velocity"]].head(3)
```

## A trustworthy rating: shrink small samples toward the mean


```python
def bayesian_rating(frame: pd.DataFrame,
                     rating_col: str,
                     votes_col: str,
                     prior_votes: int = 500) -> pd.Series:
    """Shrink each rating toward the dataset mean based on vote count.

    weighted_rating = (v / (v + m)) * R + (m / (v + m)) * C

    where R is the title's own rating, v its vote count, C the dataset-wide
    mean rating, and m the `prior_votes` "confidence" threshold: titles with
    v << m get pulled hard toward C, titles with v >> m barely move.
    """
    v = frame[votes_col]
    R = frame[rating_col]
    C = frame[rating_col].mean()
    m = prior_votes
    return ((v / (v + m)) * R + (m / (v + m)) * C).rename(f"{rating_col}_shrunk")


df["imdb_rating_shrunk"] = bayesian_rating(df, "imdb_averageRating", "imdb_numVotes")
df[["netflix_title", "imdb_averageRating", "imdb_numVotes", "imdb_rating_shrunk"]].dropna().head(3)
```

## Critic–audience alignment gap


```python
def audience_alignment_gap(frame: pd.DataFrame,
                            col_a: str = "tmdb_vote_average",
                            col_b: str = "imdb_averageRating") -> pd.Series:
    """Absolute difference between two rating sources on the same scale."""
    return (frame[col_a] - frame[col_b]).abs().rename("audience_alignment_gap")


df["audience_alignment_gap"] = audience_alignment_gap(df)
df[["netflix_title", "tmdb_vote_average", "imdb_averageRating", "audience_alignment_gap"]].dropna().head(3)
```

## Buzz volume: tame skewed vote counts with a log transform


```python
def log_buzz(frame: pd.DataFrame, votes_col: str) -> pd.Series:
    """Log1p-transformed vote count -- a compressed 'attention volume' feature."""
    return np.log1p(frame[votes_col]).rename(f"{votes_col}_log")


df["imdb_buzz_log"] = log_buzz(df, "imdb_numVotes")
df[["netflix_title", "imdb_numVotes", "imdb_buzz_log"]].dropna().head(3)
```

## Assemble everything into one feature matrix


```python
def build_feature_matrix(raw: pd.DataFrame) -> pd.DataFrame:
    """Turn the raw composite dataset into a model-ready feature matrix.

    Chains every feature function above. Rows with no IMDb match keep
    NaN for IMDb-derived features -- decide downstream whether to drop,
    impute, or model that missingness explicitly.
    """
    frame = raw.copy()
    frame["is_hit"] = label_hits(frame)
    frame["binge_velocity"] = binge_velocity(frame)
    frame["imdb_rating_shrunk"] = bayesian_rating(frame, "imdb_averageRating", "imdb_numVotes")
    frame["audience_alignment_gap"] = audience_alignment_gap(frame)
    frame["imdb_buzz_log"] = log_buzz(frame, "imdb_numVotes")

    feature_columns = [
        "binge_velocity",
        "imdb_rating_shrunk",
        "audience_alignment_gap",
        "imdb_buzz_log",
        "tmdb_popularity"
    ]
    return frame[["netflix_title", "is_hit", *feature_columns]]


features = build_feature_matrix(df)
features.head(10)
# features.describe()
```

## Exercise 9 — The Perfect Show Profile

Write your ideal show recipe using evidence from the dataset.

In Exercise 10, we built a custom mathematical formula called **`hit_score`** to evaluate and rank how close each TV series is to our "Perfect Show Profile". Here is the step-by-step logic behind this design:

#### 1. Why do we need `norm(c)` (Normalization)?
* **The Problem:** Our features are measured on completely different scales. For example, `binge_velocity` is measured in **tens of millions of hours**, while `imdb_rating_shrunk` is on a scale from **1 to 10**. If we simply add them together, the massive viewing hours will completely drown out the ratings.
* **The Solution:** We use `norm(c)` to scale every single feature to a range between **0 and 1** (where 0 is the worst in the dataset, and 1 is the best). This allows us to make a fair, "apples-to-apples" comparison.

#### 2. Why do we multiply them by "Weights" (加乘)?
* **The Reason:** Not all features are created equal. Our correlation analysis in Exercise 9 proved that some factors have a much stronger impact on making a show a "hit" than others.
* **How it works:** We assign a **weight** to each normalized feature based on its correlation strength with `is_hit`:
  * **`binge_velocity` (Weight: 0.51):** Gets the highest weight because "how fast people binge" is the absolute king of success.
  * **`audience_alignment_gap` (Weight: -0.23):** Gets a **negative** weight because a larger gap (more polarization) hurts the show. Disagreement acts as a penalty, pulling the score down.
  * **Other features (TMDb popularity, IMDb rating, buzz):** Get moderate positive weights matching their statistical impact.

#### 3. The Final Formula
$$\text{Hit Score} = \sum (\text{Weight} \times \text{Normalized Feature})$$

By multiplying each standardized feature by its importance (weight) and summing them up, we get a single, unified ranking metric. The higher the score, the more "perfect" the show is!

## Exercise 10 — Build a Hit Score

```python
def norm(c):
    return (c - c.min()) / (c.max() - c.min())

weights = {
    "binge_velocity":        0.51,
    "tmdb_popularity":       0.31,
    "imdb_rating_shrunk":    0.23,
    "audience_alignment_gap": -0.23,   # 負號：分歧大 → 扣分
    "imdb_buzz_log":         0.22,
}

df["hit_score"] = sum(w * norm(df[f]) for f, w in weights.items())
df.sort_values("hit_score", ascending=False).head(10)
```
